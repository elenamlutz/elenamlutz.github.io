```python
def dis_color(democrats):
    if (democrats == 'Washington') or (democrats == 'Oregon') or (democrats == 'Nevada') or (democrats == 'California') or (democrats == 'Colorado') or (democrats == 'New Mexico') or (democrats == 'Hawaii') or (democrats == 'Minnesota') or (democrats == 'Illinois') or (democrats == 'New York') or (democrats == 'Virginia') or (democrats == 'Maryland') or (democrats == 'New Jersey') or (democrats == 'Connecticut') or (democrats == 'Massachussetts') or (democrats == 'Vermont') or (democrats == 'Maine'):
        return '#67a9cf'
    else:
        return '#ca0020'
```

```python
def dis_party(democrats):
    if (democrats == 'Washington') or (democrats == 'Oregon') or (democrats == 'Nevada') or (democrats == 'California') or (democrats == 'Colorado') or (democrats == 'New Mexico') or (democrats == 'Hawaii') or (democrats == 'Minnesota') or (democrats == 'Illinois') or (democrats == 'New York') or (democrats == 'Virginia') or (democrats == 'Maryland') or (democrats == 'New Jersey') or (democrats == 'Connecticut') or (democrats == 'Massachussetts') or (democrats == 'Vermont') or (democrats == 'Maine'):
        return 'Democratic'
    else:
        return 'Republican'
```

```python
df_mapbox['properties.article'] = df_mapbox['properties.NAME'].apply(lambda x: dis_party(x))
```

```python
df_mapbox['properties.color'] = df_mapbox['properties.NAME'].apply(lambda x: dis_color(x))
```

```python
df_mapbox['properties.headline'] = df_mapbox['properties.NAME']
```

```python
df_mapbox['properties.name'] = df_mapbox['properties.NAME']
```

```python
df_mapbox.head()
```

```python
ok_json = json.loads(df_mapbox.to_json(orient='records'))
```

```python
def process_to_geojson(file):
    geo_data = {"type": "FeatureCollection", "features":[]}
    for row in file:
        this_dict = {"type": "Feature", "properties":{}, "geometry": {}}
        for key, value in row.items():
            key_names = key.split('.')
            if key_names[0] == 'geometry':
                this_dict['geometry'][key_names[1]] = value
            if str(key_names[0]) == 'properties':
                this_dict['properties'][key_names[1]] = value
        geo_data['features'].append(this_dict)
    return geo_data
```

```python
geo_format = process_to_geojson(ok_json)
```

```python
with open('geo-data.js', 'w') as outfile:
    outfile.write("var infoData = ")
#geojson output
with open('geo-data.js', 'a') as outfile:
    json.dump(geo_format, outfile)
```
